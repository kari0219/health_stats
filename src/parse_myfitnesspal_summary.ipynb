```python
import pandas as pd
import numpy as np
```

```python
# Read in csv file for food

df = pd.read_csv('../data/Nutrition-Summary-2019-05-01-to-2019-06-15_numberic_only.csv')
```

```python
df.describe()
df.info()
```

```python
df2 = df[df['Fiber'].notna()]


#Cast Calories to a float
#  https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy

pd.to_numeric(df2.loc[:,'Calories'])
print(df2.describe())
print()
df2.info()
```

```python
df2.groupby('Date').sum()[['Calories','Protein (g)']]
```

```python
df2.groupby('Date').sum()[['Calories','Protein (g)']].std()
```

```python
df2.groupby('Date').sum()[['Calories','Protein (g)']].mean()
```

```python
df3 = df2.groupby('Date').sum()
df3[df3['Calories'] > 1000][['Calories','Protein (g)']].mean()
```
